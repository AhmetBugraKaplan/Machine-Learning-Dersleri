```python
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = sns.load_dataset("diamonds")
df.head(3)
```

```python
df = pd.get_dummies(df,columns=["cut","color","clarity"],drop_first=True)
```

```python
df.head(3)
```

```python
y = df[["price"]]
x = df.drop("price",axis=1)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=19,train_size=0.70)
```

```python
lm = LinearRegression()
model = lm.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```

**Aslında beklenen durum model skorunda train verilince skorun daha yüksek çıkması fakat bu bizim istediğimiz bir şey değil. Aralarındaki fark ne kadar az olursa bizim için o kadar iyi yani o kadar az overfitting olmuş demektir**

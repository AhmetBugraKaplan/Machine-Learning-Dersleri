```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("audi.csv")
```

```python
df.head(3)
```

```python
df = df[["Year","Type","Mileage(miles)","Engine","PS","Transmission","Fuel","Number_of_Owners","Price(£)"]]
```

```python
df.head(3)
```

```python
df.columns=["yil","tip","mil","motor","ps","vites","yakit","sahip","fiyat"]
```

```python
df.head(3)
```

```python
df.motor = df.motor.str.replace("L","")
df.motor = pd.to_numeric(df.motor)
```

```python
df.head(3)
```

```python
df.info()
```

**Şu ana kadar tek sorunu yanında l yazmak olan benzin değişkeninin l sini sildik sonra nümerige dönüştürdük.**

**Şimdi amacımız boolean tipindeki değerleri numerice
dönüştürmek**

```python
df = pd.get_dummies(df,columns=["tip","vites","yakit"],drop_first=True)
```

```python
df.head(3)
```

```python
y = df[["fiyat"]]
x = df.drop("fiyat",axis=1)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.70,random_state=15)
```

```python
lm = LinearRegression()
model = lm.fit(x_train,y_train)
model.score(x_test,y_test)
```
